### Подготовка marketplace 

```python
place = pd.read_csv(
    "/home/fusion/project/serx/serx/parser/data/place.csv",
    sep="~",
)
place
```

```python
place["Магазин"][0].split("~")
```

```python
for i in range(place.shape[0]):
    if place.iloc[i]["Адрес"] is np.nan:
        place.iloc[i] = place.iloc[i]["Магазин"].split("~")
```

```python
place
```

```python
import re
# pattern = r'".([^"]*)".'
pattern = r'".([^"]*)'

for i in range(df.shape[0]):
    find_place = re.findall(pattern, place.iloc[i][0].strip('" '))

    if len(find_place) >= 1:
        # print(find_place[0])
        place["market"][i] = find_place[0]
    else:
        place["market"][i] = ""
```

```python
place.head(2)
```

```python
def comma_to_dot(field):
    replace_comma2dot = str(field).replace(",", ".")
    return float(replace_comma2dot)


place["longitude"] = place["Долгота"].apply(lambda x: comma_to_dot(x))
place["latitude"] = place["Широта"].apply(lambda x: comma_to_dot(x))

place["geo"] = place["longitude"].astype(str) + "," + place["latitude"].astype(str)
```

```python
place[["market", "longitude", "latitude", "geo"]]
```

```python
for i in range(10):
    pl = place.iloc[i]
    print(
        f"('{pl['market']}', '{pl['Адрес']}', '{pl['Регион']}', '{pl['Область']}', '{pl['geo']}'), "
    )
```
